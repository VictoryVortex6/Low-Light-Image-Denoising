B.) Use SVD on this image to represent it using the first 5 components, first 10 components and first 100 components.

```python
svd_img = img.reshape(img.shape[0],img.shape[1]*img.shape[2])
```

```python
U, s, V = np.linalg.svd(svd_img) 
```

```python
var_explained = np.round(s**2/np.sum(s**2), decimals=3)
var_explained
```

```python
num_components = 5
reconst_img_5 = np.array(U[:, :num_components]) .dot(np.diag(s[:num_components]).dot(
                np.array(V[:num_components, :])))
reconst_img_5 = reconst_img_5.reshape(img.shape[0],img.shape[1],img.shape[2])
reconst_img_5 = np.round(reconst_img_5)
reconst_img_5 = reconst_img_5.astype(int)
plt.imshow(reconst_img_5)
```

```python
num_components_2 = 10
reconst_img_10 = np.array(U[:, :num_components_2]) .dot(np.diag(s[:num_components_2]).dot(
                np.array(V[:num_components_2, :])))
reconst_img_10 = reconst_img_10.reshape(img.shape[0],img.shape[1],img.shape[2])
reconst_img_10 = np.round(reconst_img_10)
reconst_img_10 = reconst_img_10.astype(int)
plt.imshow(reconst_img_10)
```
